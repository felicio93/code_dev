```python
ds = xr.open_dataset(path + f'20250211_BeringSeaMoorings/ADCP_Full_bsp2_record_2a2e_5d93_fde8.nc')
obs = ds['v_comp_current'].sel(time=slice(start_date, end_date))
depth_obs = np.array(ds['depth'])
time_obs = np.array(ds['time'].sel(time=slice(start_date, end_date)))

#MODEL DATA
model = xr.open_dataset(f"{path}/mooring_{run}.nc")
model = model.sel(time=slice(start_date, end_date))
hY=model.horizontalVelY[:,0,:]
```

```python
node_idx=0
model_vel=model.horizontalVelY[:,node_idx,:]
model_depths=model.zCoordinates[:,node_idx,:]
```

```python
obs_depths = depth_obs
obs_data = obs.values
obs_data = obs_data/100
```

```python
obs_mask =  ~np.isnan(obs.values)
```

```python
interp_vel = np.full_like(obs_data,np.nan)
```

```python
for t in range(len(obs.time)):
    model_d = model_depths[t, :].values
    vel_d = model_vel[t, :].values

    mask = ~np.isnan(model_d) & ~np.isnan(vel_d)
    if np.sum(mask)<2:
        continue

    sorted_idx = np.argsort(model_d[mask])
    z_sorted = model_d[mask][sorted_idx]
    z_sorted = np.flip(z_sorted)*-1
    v_sorted = vel_d[mask][sorted_idx]
    v_sorted = np.flip(v_sorted)

    interp_vel[t,:] = np.interp(obs_depths, z_sorted, v_sorted, left=np.nan, right=np.nan)
```
